# Data Cleaning File 'Export40'

### Importing the data in a dataframe

```python
import pandas as pd
import glob

class DataLoader:
    def __init__(self, folder_path):
        self.folder_path = folder_path
        self.all_files = glob.glob(folder_path + '/*.csv')
        self.dfs = []
        
    def load_data(self):
        for filename in self.all_files:
            noise40 = pd.read_csv(filename, delimiter=';')
            self.dfs.append(noise40)
        self.noise40 = pd.concat(self.dfs, ignore_index=True)
        return self.noise40

    def num_observations(self):
        num_obs = self.noise40.shape[0]
        return f"The dataset has {num_obs} observations."
    
data_loader = DataLoader('C:/Users/fieuw/Desktop/export_40')

# Load the data into a dataframe
df = data_loader.load_data()

# Print the number of observations in the dataframe
print(data_loader.num_observations())
```

### Outlier detection 

Local Outlier Factor (demonstration)

! don't just use the indicated values as "outliers". The ideal hyperparamter "n_neighbors" should be chosen, using by a grid_search with cross-validation, which requires us first to split the dataset into a training and test data-set and to specify the featurers in order to run the machine learning algorithm

```python
from sklearn.neighbors import LocalOutlierFactor

# Select only the columns that contain numerical data
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Initialize the LOF model with your desired parameters
lof = LocalOutlierFactor(n_neighbors=40, contamination= 'auto')

# Fit the model on your data
lof.fit(df_numeric)

# Predict the outlier scores for each data point
scores = lof.negative_outlier_factor_

# Identify the outliers in your dataset
outliers = df[scores < lof.offset_]

# Count the number of outliers
n_outliers = (lof.negative_outlier_factor_ < -1.5).sum()
print("Number of outliers:", n_outliers)

# Create a new column called 'is_outlier' that indicates whether each observation is an outlier or not
df['is_outlier'] = scores < lof.offset_
```
